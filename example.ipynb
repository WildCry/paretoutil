```python
import paretoutils as pu
import matplotlib.pyplot as plt
```

```python
vei = pu.Stock('VEI.OL')

(vei.dividends/vei.WACC()).plot(label='Dividend cashflow price prediction')
vei.ClosingPrices('max').plot(label='Close Price')
plt.title(vei.ticker)
plt.show()
```

```python
returns = vei.ClosingPrices('max').resample('y').last().pct_change()
jensen_alpha = returns - vei.CAPM()
jensen_alpha.plot(label='Jensen alpha')
plt.axhline(0, c='r', linestyle='--')
plt.title(vei.ticker)
plt.legend()
plt.show()
```

```python
vei.interest_on_debt().plot(label='interest rate')
plt.title(vei.ticker)
plt.legend()
plt.show()
```
